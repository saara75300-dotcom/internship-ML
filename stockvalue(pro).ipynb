```python
inp=stock[['date','close','high']]
out=stock['splitFactor']
```

```python
LR=LinearRegression()
```

```python
Train_data=pd.concat([inp,out],axis=1)
train=Train_data.dropna()
```

```python
inp1=train[['date']]
out1=train['splitFactor']
```

```python
inp1 = train[['date']].copy()
out1=train['splitFactor']

inp1['date'] = pd.to_datetime(inp1['date']).dt.dayofyear
LR.fit(inp1,out1)
```

```python
# To make a prediction, you need to provide a numerical input for the date.
# For example, to predict for day 180 of the year:
# LR.predict([[180]])
```

```python
prediction_input = pd.DataFrame([[180]], columns=['date']) # Provide input as DataFrame with feature name
predicted_split_factor = LR.predict(prediction_input)
print(f"Predicted split factor for day 180: {predicted_split_factor[0]}")
```
