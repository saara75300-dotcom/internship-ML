```python
import pandas as pd
dia=pd.read_excel("/content/drive/MyDrive/Intern/Sara/dataset.csv/dataset.excel.xlsx")
dia.head()
```

```python
dia.isnull().sum()
```

```python
ind=dia[['age','mass','insu','plas']]
dep=dia['class']
```

```python
from sklearn.linear_model import LogisticRegression
Logr=LogisticRegression()
```

```python
Logr.fit(ind,dep)
```

```python
import numpy as np
age=int(input("Enter the age:"))
mass=int(input("Enter the mass:"))
insulin=int(input("Enter the insulin:"))
plasma=int(input("Enter the plasma:"))

# Reshape the input to a 2D array (1 sample, 4 features)
input_data = np.array([age, mass, insulin, plasma]).reshape(1, -1)
pred=Logr.predict(input_data)
print(pred)
```

```python
Logr.score(ind,dep)
```

```python
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import LabelEncoder

pval= Logr.predict(ind)
accuracy_score(dep,pval)
label_encoder = LabelEncoder()
dep_encoded = label_encoder.fit_transform(dep)
print("Original 'class' values:")
print(dep.head())
print("\nEncoded 'class' values:")
print(dep_encoded[:5])
decoded_dep = label_encoder.inverse_transform(dep_encoded)
print("\nDecoded 'class' values (first 5):")
print(decoded_dep[:5])
```
