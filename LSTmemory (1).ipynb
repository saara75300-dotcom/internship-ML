```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from keras.models import Sequential
from keras.layers import Dense, LSTM, Dropout
import math
from sklearn.metrics import mean_squared_error
```

```python
data=pd.read_csv('/content/drive/MyDrive/Intern/Sara/IBM LST.csv',index_col=0,parse_dates=True)
data.head()
```

```python
mytrain=data [:'2016'].iloc[:,1:2].values
mytest=data['2017':].iloc[:,1:2].values
```

```python
print(mytrain[0])
print(mytest[0])
```

```python
sc=MinMaxScaler(feature_range=(0,1))
mytrain_scaled=sc.fit_transform(mytrain)
```

```python
print(mytrain_scaled[0])
```

```python
mytrain_scaled
```

```python
len(mytrain_scaled)
```

```python
I_train=[]
O_train=[]
for i  in range(60,2769):
   I_train.append(mytrain_scaled[i-60:i,0])
   O_train.append(mytrain_scaled[i,0])
I_train = np.array(I_train)
O_train = np.array(O_train)
```

```python
I_train[0]
```

```python
O_train[0]
```

```python
I_train=np.array(I_train)
O_train=np.array(O_train)
```

```python
I_train.shape
```

```python
I_train=I_train.reshape(I_train.shape[0], I_train.shape[1], 1)
```

```python
model=Sequential()
model.add(LSTM(units=50,return_sequences=True,input_shape=(I_train.shape[1],1)))
model.add(Dropout(0.2))
model.add(LSTM(units=50,return_sequences=True))
model.add(Dropout(0.2))
model.add(LSTM(units=50,return_sequences=False))
model.add(Dropout(0.2))
model.add(Dense(units=1))
```

```python
model.compile(optimizer='adam',loss='mean_squared_error')
```

```python
model.fit(I_train,O_train,epochs=50,batch_size=32)
```

```python
model.save("StockIBM.h5")
```
