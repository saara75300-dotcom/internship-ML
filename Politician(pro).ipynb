```python
ind = dia[['state','year','candidate','forecast_prob','result']]
dep = dia['winflag']
```

```python
Logr=LogisticRegression()
```

```python
import pandas as pd
categorical_cols = ['state', 'candidate', 'result']
ind_encoded = pd.get_dummies(ind, columns=categorical_cols, drop_first=True)
Logr.fit(ind_encoded, dep)
```

```python
import pandas as pd
state_input = str(input("Enter the state: "))
year_input = int(input("Enter the year: "))
candidate_input = str(input("Enter the candidate: "))
forecast_prob_input = float(input("Enter the forecast_prob: "))
result_input = str(input("Enter the result: "))
new_input_df = pd.DataFrame([[
    state_input,
    year_input,
    candidate_input,
    forecast_prob_input,
    result_input
]], columns=['state', 'year', 'candidate', 'forecast_prob', 'result'])
new_input_encoded = pd.get_dummies(new_input_df, columns=categorical_cols, drop_first=True)
missing_cols = set(ind_encoded.columns) - set(new_input_encoded.columns)
for c in missing_cols:
    new_input_encoded[c] = 0
new_input_encoded = new_input_encoded[ind_encoded.columns]
pred = Logr.predict(new_input_encoded)
print(pred)
```

```python
Logr.score(ind_encoded,dep)
```

```python
from sklearn.metrics import accuracy_score
pval= Logr.predict(ind_encoded)
accuracy_score(dep,pval)
```
